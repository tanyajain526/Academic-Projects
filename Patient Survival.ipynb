BMI, or Body Mass Index, is a measurement that assesses an individual's body weight in relation to their height. It is commonly used as a quick and simple screening tool to categorize individuals into different weight status categories. The formula for BMI is as follows:

BMI= weight(kg)/(height(m))squared


The World Health Organization (WHO) defines the following BMI categories:

Underweight: BMI less than 18.5
Normal weight: BMI between 18.5 and 24.9
Overweight: BMI between 25 and 29.9
Obesity: BMI of 30 or greater, which is further categorized into three classes (Class I: 30-34.9, Class II: 35-39.9, Class III: 40 or greater)
It's important to note that while BMI is a useful tool for population-level assessments

### Managerial Insight: 


It turns out that people with a slightly lower BMI than what's considered average actually have a higher risk of various diseases. This goes against the usual belief that only high BMI is a concern. Understanding BMI can help design targeted interventions for individuals and create programs addressing weight loss/gain or nutritional education which may be beneficial for all the pateints to improve their BMI.People with lower BMI should be assessed continuously to detect diseases at an early stage.

```python
threshold_sysbp = 77.5
threshold_spo2 = 71.5
threshold_heartrate = 29.2

# Filter data for low systolic blood pressure, oxygen, and heart rate
low_vitals_data = df[(df['d1_sysbp_min'] <= threshold_sysbp) &
                       (df['d1_spo2_min'] <= threshold_spo2) &
                       (df['d1_heartrate_min'] <= threshold_heartrate)]

# Plotting count bar graph
plt.figure(figsize=(6, 4))
sns.countplot(
    x = 'hospital_death',
    data = low_vitals_data,
    palette = 'Blues'
)

plt.title('Patients with High Motor Apache and Low Vital Signs')
plt.xlabel('Hospital Death')
plt.ylabel('Count')
plt.show()
```

Even with a strong motor Apache score, individuals exhibiting low systolic blood pressure, oxygen levels, and heart rates face heightened mortality risks.Developing individualized care plans that consider both the motor Apache score and vital signs is necessary  in such cases While the motor Apache score provides insight into motor function, prioritizing interventions based on low systolic blood pressure, oxygen levels, and heart rates ensures a more tailored and effective approach to patient care.
